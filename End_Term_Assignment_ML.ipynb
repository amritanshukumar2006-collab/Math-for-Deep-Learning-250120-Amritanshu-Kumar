```python
import numpy as np

# Training data
X = np.array([1, 2, 3, 4], dtype=float)
y = np.array([2, 4, 6, 8], dtype=float)

#Initalizing the Parameters
w = 0.0
b = 0.0

learning_rate = 0.001 # Learning Rate is 0.001 as discussed
epochs = 10
n = len(X)

for epoch in range(epochs):
    for i in range(n):

        # Prediction for one sample
        y_pred = w * X[i] + b

        # Compute differentiation
        dw = 2 * (y_pred - y[i]) * X[i]
        db = 2 * (y_pred - y[i])

        # Upgrading the parameters
        w = w - learning_rate * dw
        b = b - learning_rate * db

    if epoch % 1 == 0:
        print(f"Epoch {epoch}: w={w:.4f}, b={b:.4f}")
```
